Perform standardization for wine dataset (`wine_data`) using only basic python, numpy and pandas (without using `StandardScaler` and sklearn at all). Implementation of function (or class) that can get dataset as input and return standardized dataset as output is preferrable, but not necessary.

### Simple version (1 point)

```python
# 1 point
def scale(X):
    return (X - X.mean(axis=0)) / X.std(axis=0, ddof=0)
```

```python
assert np.allclose(np.array(scale(wine_data)), StandardScaler().fit_transform(wine_data))
```

### Pipeline Version (2 points)

```python
# 2 points
from sklearn.base import BaseEstimator, TransformerMixin

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, copy=True, with_mean=True, with_std=True):
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std

    def fit(self, X, y=None):
        self.mean = X.mean(axis=0)
        self.std = np.std(X)
        return self

    def transform(self, X, y=None, copy=None):
        if copy:
            X = X.copy()
            
        res = X
        if self.with_mean:
            res = X - X.mean(axis=0)
        if self.with_std:
            res = res / X.std(ddof=0)

        return res
```

```python
assert np.allclose(CustomScaler().fit_transform(wine_data), StandardScaler().fit_transform(wine_data))
```

## Exercise  3 - Preprocessing Pipeline (2 points)

```python
wine_train, wine_val, wine_labels_train, wine_labels_val = train_test_split(wine_data, wine_labels, 
                                                                            test_size=0.3, random_state=42)
```

Train model (for example, `LogisticRegression(solver='liblinear', penalty='l1')` on raw `wine_train` data; then train same model after data scaling; then add feature selection (and train model again on scaled data). For each experiment all required preprocessing steps (if any) should be wrapped into sklearn pipeline.

Measure `accuracy` of all 3 approaches on `wine_val` dataset. Describe and explain results.

```python
from sklearn import metrics
```

```python
def make_model(scale=False, feature_selection=False, n_components=2):
    scaler = StandardScaler() if scale else None
    selector = PCA(n_components=n_components) if feature_selection else None
    
    return Pipeline([
        ('scaler', scaler),
        ('selector', selector),
        ('estimator', LogisticRegression(solver='liblinear', penalty='l1'))
    ])
```

```python
def fit_evaluate(clf, X_train, y_train, X_test, y_test, plot_confusion_matrix = False, title='Confusion Matrix'):
    clf.fit(X_train, y_train)
    if plot_confusion_matrix:
        disp = metrics.plot_confusion_matrix(clf, X_test, y_test, normalize='true')
        disp.figure_.suptitle(title)
        plt.show()
    
    return metrics.accuracy_score(y_pred=clf.predict(X_train), y_true=y_train), \
           metrics.accuracy_score(y_pred=clf.predict(X_test), y_true=y_test)
```

```python
model = make_model(scale=False)
score = fit_evaluate(model, wine_train, wine_labels_train, wine_val, wine_labels_val, plot_confusion_matrix=True, title='withouts scaler & feature selection')
print(f'Accuracy score (train, test): {score}')
```

```python
model_pca = []
model_scaled_pca = []

for i in range(1, wine_data.shape[1]):
    model_pca.append(
        fit_evaluate(make_model(scale=False, feature_selection=True, n_components=i), wine_train, wine_labels_train, wine_val, wine_labels_val)
    )
    model_scaled_pca.append(
        fit_evaluate(make_model(scale=True, feature_selection=True, n_components=i), wine_train, wine_labels_train, wine_val, wine_labels_val)
    )
```

```python
x = [i for i in range(1, wine_data.shape[1])]
plt.xlabel('n_components')
plt.ylabel('accuracy')
plt.plot(x, model_pca, label=['train_pca', 'test_pca'])
plt.plot(x, model_scaled_pca, label=['train_scaled_pca', 'test_scale_pca'])
plt.legend()
```

As wee see, we got best fit with n_components = 5. Also, we need to scale the data before the pca to get better results.

```python
model = make_model(scale=True, feature_selection=True, n_components=5)
score = fit_evaluate(model, wine_train, wine_labels_train, wine_val, wine_labels_val, plot_confusion_matrix=True, title='with scaler & feature selection')
print(f'Accuracy score (train, test): {score}')
```

As wee see, getting only 5 components from out data give as a better accuracy. <br>
It happens because other features are not so influential, and give us more noise than useful information about the target. <br>
We should also scale our data to ensure it features lies in approximately close ranges.

## Exercise 4 - manual PCA (5 points)
The task is to solve PCA as an optimization problem, without explicitly doing eigen value decomposition.
In the most general setting PCA is minimization of reconstruction error of a projection of given rank $q$

$$\min_{\mu, \lambda_1,\ldots, \lambda_n, \mathbf{V}_q} \sum_{i=1}^n ||x_i - \mu - \mathbf{V}_q \lambda_i||^2$$

With a number of steps that can be found here https://stats.stackexchange.com/a/10260 this task transforms to
 $$\max_{u_i} \sum_{i=1}^q u_i^T \mathbf{S} u_i$$
 where $\mathbf{S}$ is the sample covariance matrix (after standartization) and $u_1, \ldots, u_q$ are the $q$ are orthonormal columns in $\mathbf{V}_q$.
 Let us solve this optimization problem with `scipy.optimize` library.
 
 Additional 2 point are given for visualization of the results.

### PCA (3 points)

```python
wine_data, wine_labels = wine_sklearn['data'], wine_sklearn['target']
```

Find a covariance matrix of standartized data and assing it to S.

```python
scaler = StandardScaler()
S = np.cov(scaler.fit_transform(wine_data).T)
```

If your code is correct, the following assert should be Ok.

```python
assert np.allclose(np.linalg.norm(S), 5.787241159764733)
```

```python
from scipy.optimize import minimize

def objective(x):
    return - (x.T @ S @ x)
    
def norm_constraint(x):
    return np.linalg.norm(x) - 1

con1 = {'type': 'eq', 'fun': norm_constraint}

x0 = np.zeros(S.shape[0])

sol = minimize(objective, 
               x0, 
               constraints = [con1]
              )
x0 = sol.x
```

Hurray! We have first vector! Let's do another one.

```python
def orthogonality_constraint(x):
    return x @ x0

con2 = {'type': 'eq', 'fun': orthogonality_constraint}

x1 = np.zeros(S.shape[0])


sol = minimize(objective, 
               x1, 
               constraints = [con1, con2]
              )

x1 = sol.x
```

If your solution is correct, the following asserts should be Ok.

```python
assert np.allclose(x0@S@x0, 4.732436977583595)
assert np.allclose(x1@S@x1, 2.5110809296451233)
```

### Visualization (2 points)

Visualize the points after applying custom dimension reduction with 2 components.

```python
X = np.array([x0, x1])
wine_data_pca = np.array((-X @ scale(wine_data).T).T)

for l, c, m in zip(range(0, 3), ('blue', 'red', 'green'), ('^', 's', 'o')):
    plt.scatter(wine_data_pca[wine_labels == l, 0], wine_data_pca[wine_labels == l, 1], 
                color=c, label=f'class {l}', alpha=0.5, marker=m)
    
plt.title('Dataset after PCA')

plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
plt.legend(loc='upper right')
```
